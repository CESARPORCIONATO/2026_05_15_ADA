## Exercício 1: Classificar nível de experiência

Escreva uma função `classificar_nivel(anos)` que retorna o nível de um profissional a partir dos anos de experiência:

- 0 a 2 anos: `"Júnior"`
- 3 a 5 anos: `"Pleno"`
- 6 anos ou mais: `"Sênior"`
- Valor negativo: `"Inválido"`

**Exemplo:**

```
classificar_nivel(1)   ->  "Júnior"
classificar_nivel(4)   ->  "Pleno"
classificar_nivel(10)  ->  "Sênior"
classificar_nivel(-2)  ->  "Inválido"
```

````python
import numpy as np
def classificar_nivel(anos):
    # Se for apenas um número, converte para array
    arr = np.atleast_1d(anos)
    # Condições para classificação
    condicoes = [
        arr < 0,
        (arr >= 0) & (arr <= 2),
        (arr >= 3) & (arr <= 5),
        arr >= 6
    ]
    niveis = ["Inválido", "Júnior", "Pleno", "Sênior"]
    # Usa np.select para aplicar as condições
    resultado = np.select(condicoes, niveis)
    # Se a entrada era escalar, retorna string; senão retorna lista
    return resultado[0] if np.isscalar(anos) else resultado.tolist()
# Exemplos de uso:
print(classificar_nivel(1))     # Júnior
print(classificar_nivel(4))     # Pleno
print(classificar_nivel(10))    # Sênior
print(classificar_nivel(-2))    # Inválido
# Também funciona com lista/array:
print(classificar_nivel([1, 4, 10, -2]))
# Saída: ['Júnior', 'Pleno', 'Sênior', 'Inválido']
````

````python
# Testes
print(classificar_nivel(0)  == "Júnior")
print(classificar_nivel(2)  == "Júnior")
print(classificar_nivel(3)  == "Pleno")
print(classificar_nivel(5)  == "Pleno")
print(classificar_nivel(6)  == "Sênior")
print(classificar_nivel(20) == "Sênior")
print(classificar_nivel(-1) == "Inválido")
````


## Exercício 2: Resumo de salários

Escreva uma função `resumo_salarios(salarios)` que recebe uma lista de números (salários anuais em USD) e retorna um dicionário com:

- `"contagem"`: quantos salários foram submetidos
- `"media"`: média dos salários, arredondada para 2 casas decimais
- `"minimo"`: menor valor
- `"maximo"`: maior valor


**Exemplo:**

```
resumo_salarios([50000, 80000, 120000, 95000, 60000])
->  {"contagem": 5, "media": 81000.0, "minimo": 50000, "maximo": 120000}
```


````python
import numpy as np
def resumo_salarios(salarios):
    if not salarios:  # caso lista vazia
        return {"contagem": 0, "media": 0, "minimo": None, "maximo": None}
    arr = np.array(salarios)  # transforma a lista em array NumPy
    return {
        "contagem": arr.size,
        "media": round(arr.mean(), 2),
        "minimo": arr.min(),
        "maximo": arr.max()
    }
# Exemplo de uso:
print(resumo_salarios([50000, 80000, 120000, 95000, 60000]))
# Saída esperada:
# {"contagem": 5, "media": 81000.0, "minimo": 50000, "maximo": 120000}
````

Escreva uma função `filtrar_por_nivel(candidatos, nivel_desejado)` que:

- Recebe uma lista de dicionários, onde cada dicionário tem as chaves `"nome"` e `"anos_experiencia"`
- Recebe uma string com o nível desejado (`"Júnior"`, `"Pleno"` ou `"Sênior"`)
- Retorna uma lista com os nomes dos candidatos que correspondem ao nível

**Exemplo:**

```
candidatos = [
    {"nome": "Ana",   "anos_experiencia": 1},
    {"nome": "Bruno", "anos_experiencia": 4},
    {"nome": "Carla", "anos_experiencia": 7},
]
filtrar_por_nivel(candidatos, "Pleno")  ->  ["Bruno"]
```

Orientação: reaproveite a função `classificar_nivel`. List comprehension pode deixar o código mais curto.

````python
import numpy as np
def classificar_nivel_array(anos_array):
    arr = np.array(anos_array)
    condicoes = [
        arr < 0,
        (arr >= 0) & (arr <= 2),
        (arr >= 3) & (arr <= 5),
        arr >= 6
    ]
    niveis = ["Inválido", "Júnior", "Pleno", "Sênior"]
    return np.select(condicoes, niveis)
def filtrar_por_nivel(candidatos, nivel_desejado):
    # Extrai anos de experiência e nomes em arrays
    anos = np.array([c["anos_experiencia"] for c in candidatos])
    nomes = np.array([c["nome"] for c in candidatos])
    # Classifica todos de uma vez
    niveis = classificar_nivel_array(anos)
    # Filtra apenas os nomes que batem com o nível desejado
    return nomes[niveis == nivel_desejado].tolist()
# Exemplo de uso:
candidatos = [
    {"nome": "Ana",   "anos_experiencia": 1},
    {"nome": "Bruno", "anos_experiencia": 4},
    {"nome": "Carla", "anos_experiencia": 7},
]
print(filtrar_por_nivel(candidatos, "Pleno"))   # ["Bruno"]
print(filtrar_por_nivel(candidatos, "Júnior"))  # ["Ana"]
print(filtrar_por_nivel(candidatos, "Sênior"))  # ["Carla"]
````

````python
import numpy as np
def criar_candidato(nome, cargo, anos, pretensao, modalidade):
    # continua simples, retorna um dicionário
    return {
        "nome": nome,
        "cargo_desejado": cargo,
        "anos_experiencia": anos,
        "pretensao_salarial": pretensao,
        "modalidade": modalidade
    }
    pass
def apresentar_candidato(candidato):
    # usa a função classificar_nivel já vetorizada com numpy
    nivel = classificar_nivel(candidato["anos_experiencia"])
    return (f"Candidato:       {candidato['nome']}\n"
            f"Cargo desejado:  {candidato['cargo_desejado']}\n"
            f"Nível:           {nivel} ({candidato['anos_experiencia']} anos de experiência)\n"
            f"Pretensão:       USD {candidato['pretensao_salarial']:,}/ano\n"
            f"Modalidade:      {candidato['modalidade']}")
    pass
def esta_na_faixa(candidato, salario_min, salario_max):
    # usa numpy para comparação vetorizada (mesmo que seja escalar)
    return np.logical_and(salario_min <= candidato["pretensao_salarial"],
                          candidato["pretensao_salarial"] <= salario_max)
    pass
# --- Exemplo de uso ---
c1 = criar_candidato("Ana Lima", "Analista de Dados", 1, 55000, "remoto")
c2 = criar_candidato("Bruno Silva", "Engenheiro de Software", 4, 90000, "presencial")
print(apresentar_candidato(c1))
print("Está na faixa 50k-60k?", esta_na_faixa(c1, 50000, 60000))  # True
print("Está na faixa 60k-80k?", esta_na_faixa(c1, 60000, 80000))  # False
````

````python
import numpy as np
def adicionar_candidato(banco, candidato):
    # continua simples: adiciona ao banco (lista de dicionários)
    banco.append(candidato)
    return banco
    pass
def total_candidatos(banco):
    # usa numpy para contar rapidamente
    return np.size(banco)
    pass
def candidatos_por_nivel(banco, nivel):
    # extrai anos de experiência e nomes em arrays
    anos = np.array([c["anos_experiencia"] for c in banco])
    nomes = np.array([c["nome"] for c in banco])
    pass
    # classifica todos de uma vez com função vetorizada
    condicoes = [
        anos < 0,
        (anos >= 0) & (anos <= 2),
        (anos >= 3) & (anos <= 5),
        anos >= 6
    ]
    niveis = ["Inválido", "Júnior", "Pleno", "Sênior"]
    classificacoes = np.select(condicoes, niveis)
    # retorna apenas os candidatos que batem com o nível desejado
    return [banco[i] for i in np.where(classificacoes == nivel)[0]]
````

**Função 9:** `ranking_por_experiencia(banco)`

Retorna a lista de candidatos ordenada do mais experiente para o menos experiente.

**Exemplo:**

```
ranking = ranking_por_experiencia(banco)
ranking[0]["nome"]   ->  "Carla Mendes"  (8 anos)
ranking[-1]["nome"]  ->  "Ana Lima"      (1 ano)
```

````python
import numpy as np
def media_pretensao(banco):
    if not banco:
        return 0
    # transforma todas as pretensões em array NumPy
    pretensoes = np.array([c["pretensao_salarial"] for c in banco])
    return round(pretensoes.mean(), 2)
    pass
def candidatos_na_faixa(banco, salario_min, salario_max):
    # extrai pretensões e aplica filtro vetorizado
    pretensoes = np.array([c["pretensao_salarial"] for c in banco])
    indices = np.where((pretensoes >= salario_min) & (pretensoes <= salario_max))[0]
    return [banco[i] for i in indices]
    pass
def ranking_por_experiencia(banco):
    # extrai anos de experiência e ordena decrescentemente
    anos = np.array([c["anos_experiencia"] for c in banco])
    indices = np.argsort(-anos)  # ordena do maior para o menor
    return [banco[i] for i in indices]
    pass
````

## **Desafio bônus** — Validação com exceções

Escreva uma função `validar_dados_candidato(dados)` que recebe um dicionário e levanta uma exceção `ValueError` com mensagem clara em cada um destes casos:

- Falta alguma das chaves obrigatórias: `nome`, `cargo_desejado`, `anos_experiencia`, `pretensao_salarial`, `modalidade`
- `anos_experiencia` é negativo
- `pretensao_salarial` é negativa
- `modalidade` não é `"remoto"`, `"presencial"` ou `"híbrido"`

Se todos os dados forem válidos, a função retorna `True`.

**Exemplo:**

```
validar_dados_candidato({"nome": "Ana", ...})       ->  True
validar_dados_candidato({"nome": "Ana"})            ->  ValueError: chave faltando: cargo_desejado
validar_dados_candidato({..., "modalidade": "casa"}) ->  ValueError: modalidade inválida: casa
```

````python
import numpy as np
def validar_dados_candidato(dados):
    chaves_obrigatorias = ["nome", "cargo_desejado", "anos_experiencia", "pretensao_salarial", "modalidade"]
    # Verifica se todas as chaves obrigatórias estão presentes
    faltando = [ch for ch in chaves_obrigatorias if ch not in dados]
    if faltando:
        raise ValueError(f"chave faltando: {faltando[0]}")
    # Usa NumPy para validar valores numéricos
    anos = np.array(dados["anos_experiencia"], ndmin=1)
    pretensao = np.array(dados["pretensao_salarial"], ndmin=1)
    if np.any(anos < 0):
        raise ValueError("anos_experiencia não pode ser negativo")
    if np.any(pretensao < 0):
        raise ValueError("pretensao_salarial não pode ser negativa")
    # Valida modalidade
    modalidades_validas = np.array(["remoto", "presencial", "híbrido"])
    if dados["modalidade"] not in modalidades_validas:
        raise ValueError(f"modalidade inválida: {dados['modalidade']}")
    return True
    pass
````

````python
# Testes do bônus
dados_validos = {
    "nome": "Ana Lima",
    "cargo_desejado": "Analista de Dados",
    "anos_experiencia": 1,
    "pretensao_salarial": 55000,
    "modalidade": "remoto"
}
print(validar_dados_candidato(dados_validos) == True)
# Testes de erro — devem levantar ValueError
casos_invalidos = [
    {"nome": "Ana"},
    {**dados_validos, "anos_experiencia": -1},
    {**dados_validos, "pretensao_salarial": -100},
    {**dados_validos, "modalidade": "casa"},
]
for caso in casos_invalidos:
    try:
        validar_dados_candidato(caso)
        print(False, "deveria ter levantado ValueError:", caso)
    except ValueError as e:
        print(True, "ValueError:", e)
````
